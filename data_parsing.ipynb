```python
import os
import pandas as pd
from pathlib import Path

# point this at your raw dir
RAW_PATH = Path('data/logs/raw/')  

# grab every .parquet file
parquet_files = list(RAW_PATH.glob("*.parquet"))

# read ’em all into a list of DataFrames
df_list = []
for p in parquet_files:
    df = pd.read_parquet(p)
    df["mint"] = p.stem
    df_list.append(df)

# stitch into one big DataFrame
full_df = pd.concat(df_list, axis=0)

# optional: sort by timestamp index
full_df.sort_index(inplace=True)
```
